```python
covariates_holdout = ['CT SUB', 'L2-3 IT CTX-1', 'L2-3 IT CTX-2', 'L3 RSP-ACA', 
                      'L4-5 IT CTX', 'L5 IT CTX', 'L5 IT TPE-ENT', 'L5 NP CTX', 
                      'L5 PT CTX', 'L6 CT CTX', 'L6 IT CTX', 'L6b CTX', 'Lamp5',
                        'NP SUB', 'Pvalb', 'Sst', 'Vip']
seed = 1
rng = np.random.RandomState(seed)
seed_list = [rng.randint(100000) for i in range(0, len(covariates_holdout))]

heldout_perturbation_covariates = []

for covs, df in adata.obs.groupby("predicted.subclass"):
    #covs = frozenset(covs)
    print(covs)
    covs_perts = [
        x
        for x in df["Assign"].unique()
        if x != "NT_0"
    ]
    #print(covs_perts)
    if covs in covariates_holdout:
        random.seed(seed_list[covariates_holdout.index(covs)])
        heldout_perts = random.sample(
            covs_perts,
            int(0.7 * len(covs_perts)),
        )
    else:
        heldout_perts = []
    print(int(0.7 * len(covs_perts)))
    print(heldout_perts)
    heldout_perturbation_covariates.extend(
        [(perturbation, covs) for perturbation in heldout_perts]
    )

print(heldout_perturbation_covariates)
```
